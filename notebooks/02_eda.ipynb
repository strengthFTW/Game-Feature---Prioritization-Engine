```python
import pandas as pd
import matplotlib.pyplot as plt

df_clean = pd.read_csv('../data/processed/cleaned_reviews.csv')
```

```python
import matplotlib.pyplot as plt

df_clean["voted_up"].value_counts().plot(kind="bar")
plt.title("Positive vs Negative Reviews")
plt.xlabel("Recommended")
plt.ylabel("Count")
plt.show()
```

```python
df_clean["playtime_forever"].describe()
```

```python
playtime = df_clean[df_clean["playtime_forever"] <= df_clean["playtime_forever"].quantile(0.99)]

plt.figure(figsize=(8,5))
plt.hist(playtime["playtime_forever"], bins=50)
plt.title("Playtime Distribution (99th Percentile)")
plt.xlabel("Playtime (minutes)")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
df_clean["review_length"] = df_clean["review"].str.len()
df_clean["review_length"].describe()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df_clean["review_length"], bins=50)
plt.title("Review Length Distribution")
plt.xlabel("Characters")
plt.ylabel("Number of Reviews")
plt.show()
```

```python
import matplotlib.pyplot as plt

df_clean["created"] = pd.to_datetime(df_clean["created"])

reviews_per_month = (
    df_clean
    .set_index("created")
    .resample("M")
    .size()
)

plt.figure(figsize=(12,5))
reviews_per_month.plot()
plt.title("Reviews Over Time")
plt.xlabel("Date")
plt.ylabel("Number of Reviews")
plt.grid(True)
plt.show()
```
